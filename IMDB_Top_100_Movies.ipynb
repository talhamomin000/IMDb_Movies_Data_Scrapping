```python
# importing required libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
# websitle link 
url = "https://www.imdb.com/search/title/?sort=user_rating,asc&groups=top_1000&count=100"
```

```python
# paste yuor own User-Agent value. you can get your user-agent from "WhatIsMyBrowser.com"
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
response = requests.get(url,headers=HEADERS)
response
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
#creating an empty list, so that we can append the values
movie_title = []
release_year = []
rating = []
run_time = []
votes = []
metascore = []
movie_category = []
description = []
```

```python
# storing all the information(html code) in this variable
all_movie_data = soup.findAll("li", attrs={'class':'ipc-metadata-list-summary-item'})
```

```python
# fetching one by one movie data using for loop and apending to its respective list

for movie in all_movie_data:
    
    title = movie.find("h3", class_ = "ipc-title__text").text.split('. ', 1)[-1]
    movie_title .append(title)
    
    year = movie.find('span', class_= "sc-be6f1408-8 fcCUPU dli-title-metadata-item").text
    release_year.append(year)
    
    time = movie.find("div",class_ = "sc-be6f1408-7 iUtHEN dli-title-metadata").select_one("div span:nth-of-type(2)").text
    run_time.append(time)

    movie_rating = movie.find('span', class_="ipc-rating-star--imdb").text[0:3]
    rating.append(movie_rating)

    no_of_votes = movie.find('div', class_ = "sc-f24f1c5c-0 cPpOqU").text[5:]
    votes.append(no_of_votes)
    
#     here, Try and Except method is used to handle error and null values
    try:
        meta = movie.find('span', class_ = 'sc-b0901df4-0 bcQdDJ metacritic-score-box').text
    except AttributeError:
        meta= "NA"
    metascore.append(meta)


    category = movie.find("div",class_ = "sc-be6f1408-7 iUtHEN dli-title-metadata").select_one("div span:nth-of-type(3)").text
    movie_category.append(category)

#creating a dataframe using pandas library
movies_DF = pd.DataFrame({'Name of Movie':movie_title, 'Release Year': release_year, 'Run Time': run_time, 'Rating':rating,
                         'Total Votes':votes, 'Metascore':metascore,'Category':movie_category})


```
